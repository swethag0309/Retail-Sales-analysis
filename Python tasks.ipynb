```python
from datetime import date
from datetime import datetime, timedelta
```

```python
date_offset = orders.order_date.max() +timedelta(days= 1)
```

```python
lastdate = co.groupby(['customer_id','first_name'])['order_date'].max()
```

```python
date_offset = pd.to_datetime(date_offset)
```

```python
lastdate = pd.to_datetime(lastdate)
```

```python
recency = (date_offset - lastdate).dt.days.reset_index()
recency
```

```python
Frequency = ooi.groupby('customer_id')['order_id'].agg(['count']).reset_index()
Frequency
```

```python
Monetary = ooi.groupby('customer_id')['Total_Price'].agg(['sum']).reset_index()
Monetary
```

```python
rfm= pd.merge(recency,Frequency, on= 'customer_id')
rfm
```

```python
RFM = pd.merge(rfm, Monetary, on = 'customer_id')
RFM.columns = ['customer_id', 'first_name', 'Recency', 'Frequency', 'Monetary']
RFM
```

```python
import sqlite3
conn = sqlite3.connect('retail_sales')
RFM.to_sql('customer_segment', engine, if_exists = "replace", index = False)
conn.close()
```
